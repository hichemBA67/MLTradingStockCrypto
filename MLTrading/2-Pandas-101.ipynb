# Imports

```python
import pandas as pd
import numpy as np
from pandas_datareader.data import DataReader
import matplotlib.pyplot as plt
```

# Extract and view data


```python
start_date = "2017-01-01"
end_date = "2022-10-14"
symbol = "ETH-USD"

df = DataReader(name=symbol, data_source='yahoo', start=start_date, end=end_date)
df
```

# Working with returns

```python
# Add column
df["Returns"] = df["Close"].pct_change()
df.head(3)
```

```python
# Drop Na
df.dropna(inplace=True)
```

```python
# Calculate log returns
df["lrets"] = np.log(df["Close"] / df["Close"].shift(1))
df.head(3)
```

```python
# CumSum log returns
df["lrets_cumsum"] = df["lrets"].cumsum()
df.head(3)
```

```python
# Normalize Log returns
df["rets_normal"] = np.exp(df["lrets_cumsum"]) - 1
df.head()
```

```python
# Fill na
df = df.fillna(0)
df
```

# Structure Changes

```python
# Create a copy of DataFrame
df_new = df.copy()
df_new.head()
```

```python
# Drop columns
df_new.drop(columns=["High", "Low", "Adj Close"], inplace=True)
df_new.head()
```

```python
# Work with certain rows and columns
df_new.iloc[1:3,2:]
```

# Conditionals

```python
# Add conditonal statement
df_new.loc[df["Close"].shift(-1) > df["Close"], "TARGET"] = 1
df_new.loc[df["Close"].shift(-1) <= df["Close"], "TARGET"] = -1

df_new.head()
```

# Iterations

```python
# Iterate over data frame
i = 0
for index, row in df_new.iterrows():
    print(index, row["Close"], row["TARGET"])
    if i >= 4:
        break
    i+=1
```
